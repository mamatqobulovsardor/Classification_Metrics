##Klassifikatsiya metrikalari

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

##Ma'lumotlarni yuklash

```python
df = pd.read_csv("https://raw.githubusercontent.com/PlayingNumbers/ML_Process_Course/refs/heads/main/12_model_evaluation/ifood_df.csv")
```

```python
df
```

##Muammo tavsifi
##Ma'lumotlarni qayta ishlash

```python
cols = ['AcceptedCmp1',
       'AcceptedCmp2',
       'AcceptedCmp3',
       'AcceptedCmp4',
       'AcceptedCmp5',
       'AcceptedCmp']

df['AcceptedCmp'] = df['AcceptedCmp1'] + df['AcceptedCmp2'] + df['AcceptedCmp3'] + df['AcceptedCmp4'] + df['AcceptedCmp5']+ df['Response']

df['AcceptedCmp'] = np.where(df['AcceptedCmp'] > 0, 1, 0)
```

```python
df[df['AcceptedCmp'] == 0]
```

##Kross-valyidatsiya

```python
from sklearn.model_selection import train_test_split

y = df['AcceptedCmp']
X = df.drop(cols,axis=1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

##Model

```python
from sklearn.linear_model import LogisticRegression

rf = LogisticRegression()

rf.fit(X_train, y_train)
y_preds = rf.predict(X_test)
```

##Baholash  
###To'g'rilik, Aniqlik, Qaytarish

```python
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
)


accuracy = accuracy_score(y_test, y_preds)
precision = precision_score(y_test, y_preds)
recall = recall_score(y_test, y_preds)


print("Accuracy: {0}".format(accuracy))
print("Precision: {0}".format(precision))
print("Recall: {0}".format(recall))
```

##F1-Score

```python
f1 = f1_score(y_test, y_preds)
print("F1: {0}".format(f1))
```

##ROC-AUC, PR-AUC

```python
from sklearn.metrics import (
    roc_auc_score,
    average_precision_score
)


roc_auc = roc_auc_score(y_test, y_preds)

print("ROC-AUC: {0}".format(roc_auc))
```

```python
from sklearn.metrics import (
    average_precision_score
)

pr_auc = average_precision_score(y_test, y_preds)

print("PR-AUC: {0}".format(pr_auc))
```

##Log Loss

```python
from sklearn.metrics import (
    log_loss
)

log_loss = log_loss(y_test, y_preds)

print("Log Loss: {0}".format(log_loss))
```
